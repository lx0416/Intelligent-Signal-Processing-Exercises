```python
from thinkdsp import read_wave
soundfile4 = read_wave("Ex3_sound4.wav")
soundfile4.make_spectrum().plot()
```

Comparing the 3 sounds, sound4 is the only one that is suspicious. Since there is a variation at approximately 19000, we use the value to generate the carrier signal using cosine.

```python
dur4 = soundfile4.duration
frameRate4 = soundfile4.framerate

carsig = CosSignal(freq=19000)
wave4 = carsig.make_wave(duration=dur4, framerate=frameRate4)
```

```python
message = soundfile4*wave4
message.make_audio()
```

From the audio generated, we can hear the hidden 4-digit code, which is 1891
