# Functions to check format

Some functions will check for the line of string that contains the digital format required to be extracted. These include the video formats, which is contained within "Stream #0:0" and audio formats contained within "Stream #0:1", as can be seen above and below. The functions can then make use of this and check if the digital formats are accepted. 

```python
# testing for video format container
def testVideoFormat(string, file):
    if "mp4" in string:
        print("Video format OK!")
        file.write("Video format OK!" + "\n")
    else:
        print("Video format not correct")
        file.write("Video format not correct" + "\n")

# testing for audio format aac
def testAudioCodec(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:1" in test:
            if "aac" not in test:
                print("Audio codec not correct")
                file.write("Audio codec not correct" + "\n")
            else:
                print("Audio codec OK!")
                file.write("Audio codec OK!" + "\n")
                
# testing for audio channels stereo
def testAudioChannels(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:1" in test:
            if "stereo" or "2 channels" in test: # since stereo can also be known to have 2 channels
                print("Audio channels OK!")
                file.write("Audio channels OK!" + "\n")
            else:
                print("Audio channels not correct")
                file.write("Audio channels not correct" + "\n")
            
# testing for video codex h264l, aspect ratio 16:9
def testVideoCodec(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:0" in test:
            if "h264" not in test:
                print("Video codec not correct")
                file.write("Video codec not correct" + "\n")
            else:
                print("Video codec OK!")
                file.write("Video codec OK!" + "\n")

# testing for frame rate 25 fps
def testFrameRate(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:0" in test:
            if "25 fps" not in test:
                print("Frame rate not correct")
                file.write("Frame rate not correct" + "\n")
            else:
                print("Frame rate OK!")
                file.write("Frame rate OK!" + "\n")
                
# testing for aspect ratio 16:9
def testAspectRatio(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:0" in test:
            if "16:9" in test:
                print("Aspect ratio OK!")
                file.write("Aspect ratio OK!" + "\n")
            else:
                print("Aspect ratio not correct")
                file.write("Aspect ratio not correct" + "\n")
                
# testing for video resolution 640x360
def testResolution(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:0" in test:
            if "640x360" in test:
                print("Resolution OK!")
                file.write("Resolution OK!" + "\n")
            else:
                print("Resolution not correct")
                file.write("Resolution not correct" + "\n")
            
# testing for video bit rate
def testVideoBitRate(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:0" in test:
            sections = test.split(",")
            for s in sections:
                s = s.strip()
                if "kb/s" in s:
                    kbs = s.split("kb/s")                    
                    kbs = kbs[0].strip()
    # since we require this digital format to be compared in mb/s, we convert it from kb/s
    kbsint = int(kbs)
    mbs = kbsint/1000
    if mbs < 2 or mbs > 5:
        print("Video bit rate not correct")
        file.write("Video bit rate not correct" + "\n")
    else:
        print("Video bit rate OK!")
        file.write("Video bit rate OK!" + "\n")
        
# testing for audio bit rate 256kb/s
def testAudioBitRate(op, file):
    for i in op:
        test = i.strip()
        if "Stream #0:1" in test:
            sections = test.split(",")
            for s in sections:
                s = s.strip()
                if "kb/s" in s:
                    kbs = s.split("kb/s")                  
                    kbs = kbs[0].strip()
    kbsint = int(kbs)
    if kbsint > 256:
        print("Audio bit rate not correct")
        file.write("Audio bit rate not correct" + "\n")
    else:
        print("Audio bit rate OK!")
        file.write("Audio bit rate OK!" + "\n")

# compiling all functions into one to generate report.txt
def testing(string, op, file):
    testVideoFormat(string, file)
    testVideoCodec(op, file)
    testAudioCodec(op, file)
    testFrameRate(op, file)
    testAspectRatio(op, file)
    testResolution(op, file)
    testVideoBitRate(op, file)
    testAudioBitRate(op, file)
    testAudioChannels(op, file)
    file.write("-------------------------------" + "\n")
```

# Conversion function

Function created to automate conversion of incorrect formats

```python
# converts files to the correct format
def convert(inputfilepath):
    # splitting the input file path to direct the new folder
    slash = inputfilepath.split('/')
    outputfolder = slash[0] + "_formatOK/" + slash[1]
    outputfilepath = outputfolder[:-4] + "_formatOK" + ".mp4"
    print(outputfilepath)
    # conversion
    !ffmpeg -hide_banner -y -i $inputfilepath -c:v h264 -c:a aac -b:v 3M -aspect 16:9 -ac 2 -r 25 -s 640x360 -b:a 256k $outputfilepath
```

# Automated generation of report in TXT and conversion

This step reads in the files and creates a TXT file called report.txt which specifies which fields they do or do not adhere to. Next, the convert function is called to convert the films into the digital format specified, and the new files are saved under the folder named "Exercise3_Films_formatOK".

```python
import os

textfile = open("report.txt", "w")
textfile.write("-------------------------------" + "\n")
directory = 'Exercise3_Films'
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(filepath):
        # getting just the file
        purefile = filepath.strip("Exercise3_Films/")
        print(purefile)
        # probing the file for all digital formats
        op = !ffprobe -hide_banner $filepath
        # converting the file
        convert(filepath)
        # writing to TXT file
        textfile.write(purefile + "\n" + "-------------------------------" + "\n")
        testing(purefile, op, textfile)
        print("-------------------------------")
# closing the file
textfile.close()
```

# Final result

The final report is saved in the TXT file final.txt, which sllows the user to make sure that the digital format after conversion is correct.

```python
import os

textfile = open("final.txt", "w")
textfile.write("-------------------------------" + "\n")
directory = 'Exercise3_Films_formatOK'
for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(filepath):
        purefile = filepath.strip("Exercise3_Films_formatOK/")
        print(purefile)
        op = !ffprobe -hide_banner $filepath
        textfile.write(purefile + "\n" + "-------------------------------" + "\n")
        testing(purefile, op, textfile)
        print("-------------------------------")
textfile.close()
```
